```python
# dropping few columns and make a new filtered data set

kashti2 = kashti.drop(columns=['deck', 'embark_town', 'alone'])
kashti2.head()
```

```python
kashti2.mean()
```

```python
kashti2.value_counts(["survived"])
```

```python
kashti2.value_counts(["survived"]).mean()
```

```python
kashti2.groupby(["sex"]).mean()
```

```python
kashti2.groupby(["sex", "class"]).mean()
```

```python
# to check how many children survived 

kashti2[kashti2 ['age'] < 18].mean()
```

```python
# grouping the children among girl and boys

kashti2[kashti2 ['age'] < 18].groupby(["sex", "class"]).mean()
```
