## - Adding Limits to plots

```python
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset("iris")
p1 = sns.lineplot(x="sepal_length", y="sepal_width", data=iris)

# below is the function we can use to set limits for x and y: 
plt.xlim(5)
plt.ylim(3)

p1.set_title("Line Plot")
plt.show()
```

## - Adding Styles

```python
import seaborn as sns
import matplotlib.pyplot as plt

# first you need to make style=None 
sns.set_style(style=None, rc=None)

iris = sns.load_dataset("iris")
p1 = sns.lineplot(x="sepal_length", y="sepal_width", data=iris)
p1.set_title("Line Plot")

# then you set the style using the function below:
sns.set_style("dark")

plt.show()
```

## - Size of figure

```python
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset("iris")

# change figure size using the function below:
plt.figure(figsize=(4, 2))

p1 = sns.lineplot(x="sepal_length", y="sepal_width", data=iris)
p1.set_title("Line Plot")
plt.show()
```

## - Changing Order in Bar Plots

```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset("titanic")

# to change the order we simply add the following thing (order=["female", "male"])
p1 = sns.barplot(x="sex", y="alone", hue="who", data=titanic, order=["female", "male"])

p1.set_title("Bar Plot")
plt.show()
```

## - Changing color of bars

```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset("titanic")

# to change the color of bars we simple add the following thing color="red/anycolor"
p1 = sns.barplot(x="sex", y="alone", hue="who", data=titanic, order=["female", "male"], color="yellow")

p1.set_title("Bar Plot")
plt.show()
```

## - Removing the black lines or bars

```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset("titanic")

# to remove the black lines we simply add ci=None
p1 = sns.barplot(x="sex", y="alone", hue="who", data=titanic, order=["female", "male"], color="yellow",ci=None)

p1.set_title("Bar Plot")
plt.show()
```

## - Changing color palettes

```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset("titanic")

# to change the color palettes we simply add palette="your_choice_of_palette"
p1 = sns.barplot(x="sex", y="alone", hue="who", data=titanic, order=["female", "male"], color="yellow",ci=None, palette="dark")

p1.set_title("Bar Plot")
plt.show()
```

## - Advance Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset("titanic")

sns.barplot(x="class", y="fare", data=titanic,
            linewidth=3, facecolor=(1, 1, 1, 0),
            errcolor=".2", edgecolor=".2")
```

## - Box Plot

```python
import seaborn

seaborn.set(style='whitegrid')
titanic = seaborn.load_dataset("titanic")

seaborn.boxplot(x="class",
                y="fare",
                data=titanic)
```

```python
# using tips data set we create a box plot
import seaborn
seaborn.set(style='whitegrid')

tips = seaborn.load_dataset("tips")

seaborn.boxplot(x="day",
                y="tip",
                data=tips)
```
